```python
# ============================================================
# Cell 5: GWAS reader and automatic column detection
# ============================================================

import pandas as pd
import numpy as np
import gzip
import re
from pathlib import Path

def infer_separator(file_path: Path):
    name = file_path.name.lower()
    if name.endswith(".csv") or name.endswith(".csv.gz"):
        return ","
    return "\t"

def read_gwas_file(file_path: Path, nrows=None):
    sep = infer_separator(file_path)
    return pd.read_csv(
        file_path,
        sep=sep,
        compression="infer",
        nrows=nrows,
        low_memory=False
    )

COLUMN_CANDIDATES = {
    "SNP": [
        "SNP", "rsid", "rsID", "RSID", "variant_id", "variant", "MarkerName",
        "marker", "ID", "hm_rsid"
    ],
    "CHR": [
        "CHR", "CHROM", "chromosome", "chrom", "hm_chrom", "seqnames"
    ],
    "POS": [
        "POS", "BP", "base_pair_location", "position", "hm_pos", "pos",
        "variant_pos", "base_pair_position"
    ],
    "EA": [
        "EA", "effect_allele", "effectAllele", "A1", "ALT", "alt",
        "tested_allele", "hm_effect_allele"
    ],
    "OA": [
        "OA", "other_allele", "otherAllele", "A2", "REF", "ref",
        "non_effect_allele", "hm_other_allele"
    ],
    "BETA": [
        "BETA", "beta", "effect_size", "Effect", "effect", "estimate",
        "log_odds", "logOR"
    ],
    "OR": [
        "odds_ratio", "OR", "or", "oddsratio", "OddsRatio"
    ],
    "SE": [
        "SE", "se", "standard_error", "standard_error_beta",
        "stderr", "StdErr", "sebeta"
    ],
    "EAF": [
        "EAF", "eaf", "effect_allele_frequency", "A1FREQ",
        "AF", "af", "freq", "hm_effect_allele_frequency"
    ],
    "P": [
        "P", "p", "pval", "p_value", "p-value", "PVAL",
        "pvalue", "P_VALUE", "PVALUE"
    ],
    "N": [
        "N", "n", "samplesize", "sample_size", "n_samples",
        "N_total", "total_sample_size"
    ],
}

def find_column(columns, candidates):
    lower_map = {str(c).lower(): c for c in columns}
    for cand in candidates:
        if cand.lower() in lower_map:
            return lower_map[cand.lower()]
    return None

def detect_columns(df: pd.DataFrame):
    detected = {}
    for standard_name, candidates in COLUMN_CANDIDATES.items():
        detected[standard_name] = find_column(df.columns, candidates)
    return detected

def clean_chr_value(x):
    if pd.isna(x):
        return np.nan
    x = str(x).replace("chr", "").replace("CHR", "").strip()
    if x in ["X", "x"]:
        return 23
    if x in ["Y", "y"]:
        return 24
    try:
        return int(float(x))
    except:
        return np.nan

def infer_build_from_filename(file_path: Path):
    name = file_path.name.lower()
    if "grch38" in name or "hg38" in name or "build38" in name:
        return "GRCh38"
    if "grch37" in name or "hg19" in name or "build37" in name:
        return "GRCh37"
    return "Unknown"

print("GWAS helper functions are ready.")
```

```python
# ============================================================
# Cell 6: Standardise GWAS files into MR-ready columns, v2
# Supports fallback pseudo SNP ID when rsID is missing
# ============================================================

ALLOW_PSEUDO_SNP_IF_RSID_MISSING = True

def standardise_gwas_file(
    input_path: Path,
    output_path: Path,
    trait: str,
    role: str,
    force_binary_outcome_or_to_beta: bool = False,
    allow_pseudo_snp: bool = ALLOW_PSEUDO_SNP_IF_RSID_MISSING
):
    """
    Standardise GWAS summary statistics into columns:
    SNP, CHR, POS, EA, OA, BETA, SE, EAF, P, N, trait, role, source_file, build.

    If rsID/SNP is missing but CHR/POS/EA/OA exist, this version can create a pseudo SNP ID:
    CHR:POS:EA:OA

    Important:
    - Real rsID is preferred for formal MR and remote LD clumping.
    - Pseudo SNP ID is mainly useful for backend testing or when outcome uses the same coordinate-based variant ID.
    """

    print(f"\nProcessing {role}: {trait}")
    print("Input:", input_path)

    df = read_gwas_file(input_path)
    detected = detect_columns(df)

    print("Detected columns:")
    for k, v in detected.items():
        print(f"  {k}: {v}")

    # BETA can be directly available, or derived from OR
    has_beta = detected.get("BETA") is not None
    has_or = detected.get("OR") is not None

    if not has_beta and not has_or:
        raise ValueError(
            f"No BETA or odds_ratio column found for {input_path.name}. "
            "MR requires an effect size."
        )

    # Required fields excluding SNP first.
    # SNP has fallback logic below.
    required_basic = ["EA", "OA", "SE", "P"]
    missing_basic = [k for k in required_basic if detected.get(k) is None]

    if missing_basic:
        raise ValueError(f"Missing required columns for {input_path.name}: {missing_basic}")

    out = pd.DataFrame()

    # CHR and POS
    if detected.get("CHR") is not None:
        out["CHR"] = df[detected["CHR"]].apply(clean_chr_value)
    else:
        out["CHR"] = np.nan

    if detected.get("POS") is not None:
        out["POS"] = pd.to_numeric(df[detected["POS"]], errors="coerce")
    else:
        out["POS"] = np.nan

    # Alleles
    out["EA"] = df[detected["EA"]].astype(str).str.upper().str.strip()
    out["OA"] = df[detected["OA"]].astype(str).str.upper().str.strip()

    # SNP / rsID handling
    if detected.get("SNP") is not None:
        out["SNP"] = df[detected["SNP"]].astype(str).str.strip()
        snp_source_type = "detected_snp_column"

    else:
        if allow_pseudo_snp and detected.get("CHR") is not None and detected.get("POS") is not None:
            print(
                "WARNING: No rsID/SNP column detected. "
                "Creating pseudo SNP ID using CHR:POS:EA:OA."
            )

            out["SNP"] = (
                out["CHR"].astype("Int64").astype(str)
                + ":"
                + out["POS"].astype("Int64").astype(str)
                + ":"
                + out["EA"].astype(str)
                + ":"
                + out["OA"].astype(str)
            )

            snp_source_type = "pseudo_chr_pos_allele_id"

        else:
            raise ValueError(
                f"Missing SNP/rsID column for {input_path.name}, and pseudo SNP fallback is not possible."
            )

    # Effect size handling
    if has_beta:
        out["BETA"] = pd.to_numeric(df[detected["BETA"]], errors="coerce")
        effect_size_type = "beta"
    else:
        out["BETA"] = np.log(pd.to_numeric(df[detected["OR"]], errors="coerce"))
        effect_size_type = "odds_ratio_converted_to_beta"

    # For binary outcome, force OR-to-beta if odds_ratio exists
    if force_binary_outcome_or_to_beta and has_or:
        out["BETA"] = np.log(pd.to_numeric(df[detected["OR"]], errors="coerce"))
        effect_size_type = "odds_ratio_converted_to_beta"

    out["SE"] = pd.to_numeric(df[detected["SE"]], errors="coerce")

    if detected.get("EAF") is not None:
        out["EAF"] = pd.to_numeric(df[detected["EAF"]], errors="coerce")
    else:
        out["EAF"] = np.nan

    out["P"] = pd.to_numeric(df[detected["P"]], errors="coerce")

    if detected.get("N") is not None:
        out["N"] = pd.to_numeric(df[detected["N"]], errors="coerce")
    else:
        out["N"] = np.nan

    out["trait"] = trait
    out["role"] = role
    out["source_file"] = str(input_path)
    out["build"] = infer_build_from_filename(input_path)
    out["effect_size_type"] = effect_size_type
    out["snp_source_type"] = snp_source_type

    # Basic cleaning
    out = out.replace([np.inf, -np.inf], np.nan)
    out = out.dropna(subset=["SNP", "EA", "OA", "BETA", "SE", "P"])

    # Remove invalid alleles
    valid_alleles = {"A", "C", "G", "T"}
    out = out[out["EA"].isin(valid_alleles) & out["OA"].isin(valid_alleles)]
    out = out[out["EA"] != out["OA"]]

    # Remove impossible p-values
    out = out[(out["P"] > 0) & (out["P"] <= 1)]

    # Remove rows with invalid pseudo CHR/POS if pseudo SNP is used
    if snp_source_type == "pseudo_chr_pos_allele_id":
        out = out.dropna(subset=["CHR", "POS"])

    output_path.parent.mkdir(parents=True, exist_ok=True)
    out.to_csv(output_path, index=False)

    print("Output:", output_path)
    print("Rows saved:", len(out))
    print("Effect size type:", effect_size_type)
    print("SNP source type:", snp_source_type)
    print("Build inferred:", infer_build_from_filename(input_path))

    if snp_source_type == "pseudo_chr_pos_allele_id":
        print(
            "NOTE: This file does not have real rsID. "
            "The pseudo SNP ID can help preprocessing continue, "
            "but formal LD clumping/MR may require rsID-based files."
        )

    return {
        "trait": trait,
        "role": role,
        "input_file": str(input_path),
        "output_file": str(output_path),
        "rows_saved": len(out),
        "effect_size_type": effect_size_type,
        "build": infer_build_from_filename(input_path),
        "snp_source_type": snp_source_type,
        "detected_columns": detected,
    }

print("Standardisation function v2 is ready.")
```

```python
# ============================================================
# Cell 6.7: Standardise screening record columns
# ============================================================

import pandas as pd
import numpy as np
from pathlib import Path

screening_record_path = PROJECT_ROOT / "artesunate_file_screening_record.csv"

if not screening_record_path.exists():
    raise FileNotFoundError(f"Screening record not found: {screening_record_path}")

screening_df_raw = pd.read_csv(screening_record_path)

print("Raw screening record shape:", screening_df_raw.shape)
print("Raw columns:")
print(list(screening_df_raw.columns))


def pick_col(df, candidates):
    """
    Pick the first existing column from candidate names.
    """
    lower_map = {str(c).lower(): c for c in df.columns}
    for cand in candidates:
        if cand.lower() in lower_map:
            return lower_map[cand.lower()]
    return None


# Possible column names from different dashboard versions
trait_col = pick_col(screening_df_raw, [
    "trait", "Trait", "confirmed_trait", "target_trait", "detected_trait"
])

role_col = pick_col(screening_df_raw, [
    "role", "Role", "confirmed_role", "file_role"
])

build_col = pick_col(screening_df_raw, [
    "build", "Build", "genome_build", "confirmed_build", "detected_build"
])

status_col = pick_col(screening_df_raw, [
    "status", "Status", "file_status", "suitability_status", "screening_status"
])

path_col = pick_col(screening_df_raw, [
    "path", "file_path", "File Path", "full_path", "absolute_path", "selected_path"
])

name_col = pick_col(screening_df_raw, [
    "file_name", "filename", "name", "File Name"
])

effect_size_type_col = pick_col(screening_df_raw, [
    "effect_size_type", "Effect Size Type", "effect_type", "detected_effect_size_type"
])

notes_col = pick_col(screening_df_raw, [
    "notes", "Notes", "manual_notes", "comment", "remarks"
])

print("\nDetected important columns:")
print("trait_col:", trait_col)
print("role_col:", role_col)
print("build_col:", build_col)
print("status_col:", status_col)
print("path_col:", path_col)
print("name_col:", name_col)
print("effect_size_type_col:", effect_size_type_col)
print("notes_col:", notes_col)

required_record_cols = {
    "trait_col": trait_col,
    "role_col": role_col,
    "build_col": build_col,
    "status_col": status_col,
    "path_col": path_col,
}

missing_record_cols = [k for k, v in required_record_cols.items() if v is None]

if missing_record_cols:
    raise ValueError(
        "Some important columns could not be detected from the screening record: "
        + str(missing_record_cols)
        + "\nPlease check the actual column names printed above."
    )

# Create a clean internal version
screening_df = screening_df_raw.copy()

screening_df["_trait"] = screening_df[trait_col].astype(str).str.strip()
screening_df["_role"] = screening_df[role_col].astype(str).str.strip().str.lower()
screening_df["_build"] = screening_df[build_col].astype(str).str.strip()
screening_df["_status"] = screening_df[status_col].astype(str).str.strip()
screening_df["_path"] = screening_df[path_col].astype(str).str.strip()

if name_col is not None:
    screening_df["_file_name"] = screening_df[name_col].astype(str).str.strip()
else:
    screening_df["_file_name"] = screening_df["_path"].apply(lambda x: Path(x).name)

if effect_size_type_col is not None:
    screening_df["_effect_size_type"] = screening_df[effect_size_type_col].astype(str).str.strip()
else:
    screening_df["_effect_size_type"] = "Unknown"

if notes_col is not None:
    screening_df["_notes"] = screening_df[notes_col].astype(str).str.strip()
else:
    screening_df["_notes"] = ""

print("\nStandardised screening record preview:")
display(screening_df[["_trait", "_role", "_build", "_status", "_effect_size_type", "_file_name", "_path"]].head(20))
```

```python
# ============================================================
# Cell 6.8: Filter usable candidate files from screening record
# ============================================================

TARGET_EXPOSURES = ["IGF1", "SHBG", "ADIPOQ", "GDF15"]
TARGET_OUTCOME = "endometrial_cancer"

PREFERRED_BUILD = "GRCh38"

# Status keywords that are considered usable
# You can adjust these based on your dashboard labels.
USABLE_STATUS_KEYWORDS = [
    "ready",
    "suitable",
    "usable",
    "pass",
    "adaptation",
    "needs adaptation",
    "ready after adaptation"
]

BAD_STATUS_KEYWORDS = [
    "not suitable",
    "unsuitable",
    "missing",
    "fail",
    "failed",
    "not ready",
    "needs review",
    "needs build harmonisation"
]


def normalise_trait_name(x):
    """
    Normalise trait labels so that slight variations can be matched.
    """
    x = str(x).lower().replace("-", "").replace("_", "").replace(" ", "")

    if "igf" in x:
        return "IGF1"
    if "shbg" in x:
        return "SHBG"
    if "adipo" in x or "adiponectin" in x:
        return "ADIPOQ"
    if "gdf15" in x or "gdf" in x:
        return "GDF15"
    if "endometrial" in x or "uterine" in x:
        return "endometrial_cancer"

    return str(x)


def is_exposure_role(x):
    x = str(x).lower()
    return "exposure" in x


def is_outcome_role(x):
    x = str(x).lower()
    return "outcome" in x


def status_score(status):
    """
    Higher score means more usable.
    """
    s = str(status).lower()

    # Bad labels first
    for bad in BAD_STATUS_KEYWORDS:
        if bad in s:
            return -100

    # Strong usable labels
    if "ready after adaptation" in s:
        return 80
    if "ready" in s:
        return 100
    if "suitable" in s:
        return 90
    if "adaptation" in s:
        return 70
    if "usable" in s:
        return 70
    if "pass" in s:
        return 60

    # Unknown but not explicitly bad
    return 0


def build_score(build):
    b = str(build).lower()
    if "grch38" in b or "hg38" in b or "build38" in b:
        return 100
    if "grch37" in b or "hg19" in b or "build37" in b:
        return 50
    return 0


def effect_size_score(effect_size_type, role):
    e = str(effect_size_type).lower()
    r = str(role).lower()

    if "beta" in e:
        return 100

    # Outcome can use odds_ratio if later converted to beta
    if "odds" in e or e == "or":
        if "outcome" in r:
            return 80
        return 30

    return 0


def path_exists_score(path):
    return 100 if Path(str(path)).exists() else -100


candidate_df = screening_df.copy()
candidate_df["_normalised_trait"] = candidate_df["_trait"].apply(normalise_trait_name)
candidate_df["_status_score"] = candidate_df["_status"].apply(status_score)
candidate_df["_build_score"] = candidate_df["_build"].apply(build_score)
candidate_df["_effect_size_score"] = candidate_df.apply(
    lambda row: effect_size_score(row["_effect_size_type"], row["_role"]),
    axis=1
)
candidate_df["_path_exists_score"] = candidate_df["_path"].apply(path_exists_score)

# Keep only target traits
target_all_traits = TARGET_EXPOSURES + [TARGET_OUTCOME]
candidate_df = candidate_df[candidate_df["_normalised_trait"].isin(target_all_traits)].copy()

# Keep role-consistent rows
candidate_df = candidate_df[
    (
        candidate_df["_normalised_trait"].isin(TARGET_EXPOSURES)
        & candidate_df["_role"].apply(is_exposure_role)
    )
    |
    (
        candidate_df["_normalised_trait"].eq(TARGET_OUTCOME)
        & candidate_df["_role"].apply(is_outcome_role)
    )
].copy()

# Overall file score
candidate_df["_overall_score"] = (
    candidate_df["_status_score"]
    + candidate_df["_build_score"]
    + candidate_df["_effect_size_score"]
    + candidate_df["_path_exists_score"]
)

# Sort by trait and score
candidate_df = candidate_df.sort_values(
    by=["_normalised_trait", "_overall_score"],
    ascending=[True, False]
)

print("Candidate files after filtering:")
display(candidate_df[[
    "_normalised_trait",
    "_role",
    "_build",
    "_status",
    "_effect_size_type",
    "_overall_score",
    "_file_name",
    "_path"
]].head(100))
```

```python
# ============================================================
# Cell 6.9: Automatically select best 4-exposure + 1-outcome analysis set
# ============================================================

def select_best_file_for_trait(candidate_df, trait, role):
    """
    Select the highest-scoring file for one trait and role.
    """
    sub = candidate_df[candidate_df["_normalised_trait"].eq(trait)].copy()

    if role == "exposure":
        sub = sub[sub["_role"].apply(is_exposure_role)]
    elif role == "outcome":
        sub = sub[sub["_role"].apply(is_outcome_role)]

    if len(sub) == 0:
        return None

    # Prefer the preferred build explicitly if available
    preferred = sub[sub["_build"].astype(str).str.contains(PREFERRED_BUILD, case=False, na=False)].copy()

    if len(preferred) > 0:
        preferred = preferred.sort_values("_overall_score", ascending=False)
        return preferred.iloc[0]

    # Otherwise use highest score
    sub = sub.sort_values("_overall_score", ascending=False)
    return sub.iloc[0]


selected_rows = []

for trait in TARGET_EXPOSURES:
    row = select_best_file_for_trait(candidate_df, trait, role="exposure")
    if row is None:
        print(f"WARNING: No usable exposure candidate found for {trait}")
    else:
        selected_rows.append(row)

outcome_row = select_best_file_for_trait(candidate_df, TARGET_OUTCOME, role="outcome")
if outcome_row is None:
    print(f"WARNING: No usable outcome candidate found for {TARGET_OUTCOME}")
else:
    selected_rows.append(outcome_row)

selected_analysis_df = pd.DataFrame(selected_rows)

print("Automatically selected analysis set:")
display(selected_analysis_df[[
    "_normalised_trait",
    "_role",
    "_build",
    "_status",
    "_effect_size_type",
    "_overall_score",
    "_file_name",
    "_path"
]])

# Build route check
selected_builds = selected_analysis_df["_build"].astype(str).unique().tolist()

if len(selected_analysis_df) < 5:
    set_status = "Needs review: missing required exposure or outcome"
elif len(set([b.lower() for b in selected_builds])) == 1:
    if any(selected_analysis_df["_status"].astype(str).str.lower().str.contains("adaptation", na=False)):
        set_status = "Ready after adaptation"
    else:
        set_status = "Ready"
else:
    set_status = "Needs build harmonisation"

print("\nSelected builds:", selected_builds)
print("Analysis set status:", set_status)

# Create SELECTED_ANALYSIS_FILES dictionary for Cell 7
SELECTED_ANALYSIS_FILES = {}

for _, row in selected_analysis_df.iterrows():
    trait = row["_normalised_trait"]
    SELECTED_ANALYSIS_FILES[trait] = row["_path"]

print("\nSELECTED_ANALYSIS_FILES:")
for trait, path in SELECTED_ANALYSIS_FILES.items():
    print(f"{trait}: {path}")

# Save selected analysis set record
auto_selected_set_path = SUMMARY_DIR / "auto_selected_analysis_set.csv"
selected_analysis_df.to_csv(auto_selected_set_path, index=False)

print("\nAuto-selected analysis set saved to:", auto_selected_set_path)
```

```python
# ============================================================
# Cell 7.2: Create slim standardised files for backend MR
# ============================================================

import pandas as pd
import shutil
from pathlib import Path

SLIM_EXPOSURE_ROOT = BACKEND_ROOT / "slim_prepared_exposures"
SLIM_OUTCOME_ROOT = BACKEND_ROOT / "slim_prepared_outcome"

if SLIM_EXPOSURE_ROOT.exists():
    shutil.rmtree(SLIM_EXPOSURE_ROOT)
if SLIM_OUTCOME_ROOT.exists():
    shutil.rmtree(SLIM_OUTCOME_ROOT)

SLIM_EXPOSURE_ROOT.mkdir(parents=True, exist_ok=True)
SLIM_OUTCOME_ROOT.mkdir(parents=True, exist_ok=True)

MR_REQUIRED_COLUMNS = ["SNP", "CHR", "POS", "EA", "OA", "BETA", "SE", "EAF", "P", "N"]

def create_slim_csv(input_file: Path, output_file: Path, chunksize=500000):
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_chunk = True
    total_rows = 0

    for chunk in pd.read_csv(input_file, chunksize=chunksize, low_memory=False):
        available_cols = [c for c in MR_REQUIRED_COLUMNS if c in chunk.columns]
        slim = chunk[available_cols].copy()

        # Ensure all required columns exist
        for col in MR_REQUIRED_COLUMNS:
            if col not in slim.columns:
                slim[col] = pd.NA

        slim = slim[MR_REQUIRED_COLUMNS]

        slim.to_csv(
            output_file,
            index=False,
            mode="w" if first_chunk else "a",
            header=first_chunk
        )

        total_rows += len(slim)
        first_chunk = False

    return total_rows

slim_records = []

# Exposures
for trait in TARGET_EXPOSURES:
    input_file = PREPARED_EXPOSURE_ROOT / trait / f"{trait}_standardised.csv"
    output_file = SLIM_EXPOSURE_ROOT / trait / f"{trait}_standardised_slim.csv"

    if input_file.exists():
        print(f"Creating slim file for {trait}...")
        rows = create_slim_csv(input_file, output_file)

        slim_records.append({
            "trait": trait,
            "role": "exposure",
            "input_file": str(input_file),
            "output_file": str(output_file),
            "rows": rows,
            "size_MB": round(output_file.stat().st_size / 1024 / 1024, 2)
        })

        print(f"Saved: {output_file}")
        print(f"Rows: {rows}, Size MB: {round(output_file.stat().st_size / 1024 / 1024, 2)}")
    else:
        print(f"WARNING: Missing prepared exposure file for {trait}: {input_file}")

# Outcome
input_file = PREPARED_OUTCOME_ROOT / "endometrial_cancer_standardised.csv"
output_file = SLIM_OUTCOME_ROOT / "endometrial_cancer_standardised_slim.csv"

if input_file.exists():
    print("Creating slim file for endometrial_cancer...")
    rows = create_slim_csv(input_file, output_file)

    slim_records.append({
        "trait": TARGET_OUTCOME,
        "role": "outcome",
        "input_file": str(input_file),
        "output_file": str(output_file),
        "rows": rows,
        "size_MB": round(output_file.stat().st_size / 1024 / 1024, 2)
    })

    print(f"Saved: {output_file}")
    print(f"Rows: {rows}, Size MB: {round(output_file.stat().st_size / 1024 / 1024, 2)}")
else:
    print("WARNING: Missing prepared outcome file.")

slim_summary = pd.DataFrame(slim_records)
slim_summary_path = SUMMARY_DIR / "slim_prepared_file_summary.csv"
slim_summary.to_csv(slim_summary_path, index=False)

print("\nSlim file summary:")
display(slim_summary)
print("Saved to:", slim_summary_path)
```

```python
# ============================================================
# Cell 8: Lightweight QC for slim prepared files
# ============================================================

import pandas as pd
import numpy as np
from pathlib import Path

def chunk_qc_csv(path: Path, chunksize=500000):
    if not path.exists():
        return {
            "file": str(path),
            "exists": False,
            "rows": 0,
            "unique_snps_estimated": np.nan,
            "min_p": np.nan,
            "num_p5e8": 0,
            "num_p1e5": 0,
            "size_MB": 0,
        }

    total_rows = 0
    min_p = np.inf
    num_p5e8 = 0
    num_p1e5 = 0

    # Exact unique SNP can be memory-heavy, so only estimate if file is small.
    unique_snps = set()
    collect_unique = path.stat().st_size < 500 * 1024 * 1024

    for chunk in pd.read_csv(path, chunksize=chunksize, low_memory=False):
        total_rows += len(chunk)

        if "P" in chunk.columns:
            p = pd.to_numeric(chunk["P"], errors="coerce")
            min_p = min(min_p, p.min(skipna=True))
            num_p5e8 += int((p < 5e-8).sum())
            num_p1e5 += int((p < 1e-5).sum())

        if collect_unique and "SNP" in chunk.columns:
            unique_snps.update(chunk["SNP"].dropna().astype(str).tolist())

    return {
        "file": str(path),
        "exists": True,
        "rows": total_rows,
        "unique_snps_estimated": len(unique_snps) if collect_unique else "Skipped_large_file",
        "min_p": min_p if min_p != np.inf else np.nan,
        "num_p5e8": num_p5e8,
        "num_p1e5": num_p1e5,
        "size_MB": round(path.stat().st_size / 1024 / 1024, 2),
    }

slim_files = []

for trait in TARGET_EXPOSURES:
    slim_files.append(SLIM_EXPOSURE_ROOT / trait / f"{trait}_standardised_slim.csv")

slim_files.append(SLIM_OUTCOME_ROOT / "endometrial_cancer_standardised_slim.csv")

slim_qc_records = []

for path in slim_files:
    print("QC:", path)
    slim_qc_records.append(chunk_qc_csv(path))

slim_qc_df = pd.DataFrame(slim_qc_records)
slim_qc_path = SUMMARY_DIR / "slim_prepared_file_qc.csv"
slim_qc_df.to_csv(slim_qc_path, index=False)

display(slim_qc_df)
print("Saved to:", slim_qc_path)
```

```python
# ============================================================
# Cell 9: Chunk-based cis significant variant selection
# ============================================================

import pandas as pd
import numpy as np
import shutil
from pathlib import Path

SIG_THRESHOLD = 5e-8
CIS_WINDOW = 300000

GENE_COORDINATES_GRCH38 = {
    "IGF1":  (12, 102395874, 102481839),
    "SHBG":  (17, 7618495, 7632902),
    "ADIPOQ": (3, 186842033, 186858463),
    "GDF15": (19, 18374742, 18389181),
}

if SIG_VARIANT_ROOT.exists():
    shutil.rmtree(SIG_VARIANT_ROOT)
SIG_VARIANT_ROOT.mkdir(parents=True, exist_ok=True)

def select_cis_significant_variants_chunked(
    trait,
    input_file: Path,
    output_dir: Path,
    gene_coordinates,
    window=300000,
    p_threshold=5e-8,
    chunksize=500000
):
    chrom, start, end = gene_coordinates
    lower = start - window
    upper = end + window

    selected_chunks = []
    total_rows = 0
    cis_rows = 0
    sig_rows = 0

    print(f"\nSelecting cis significant variants for {trait}")
    print("Input:", input_file)
    print("CHR:", chrom, "Region:", lower, "-", upper)
    print("P threshold:", p_threshold)

    for chunk in pd.read_csv(input_file, chunksize=chunksize, low_memory=False):
        total_rows += len(chunk)

        chunk["CHR"] = pd.to_numeric(chunk["CHR"], errors="coerce")
        chunk["POS"] = pd.to_numeric(chunk["POS"], errors="coerce")
        chunk["P"] = pd.to_numeric(chunk["P"], errors="coerce")

        cis = chunk[
            (chunk["CHR"] == chrom)
            & (chunk["POS"] >= lower)
            & (chunk["POS"] <= upper)
        ].copy()

        cis_rows += len(cis)

        sig = cis[cis["P"] < p_threshold].copy()
        sig_rows += len(sig)

        if len(sig) > 0:
            selected_chunks.append(sig)

    if len(selected_chunks) > 0:
        out_df = pd.concat(selected_chunks, ignore_index=True)
        out_df = out_df.sort_values("P")

        output_file = output_dir / f"{trait}_full_sig_variants.csv"
        out_df.to_csv(output_file, index=False)

        print("Saved:", output_file)
        print("Total rows:", total_rows)
        print("Cis rows:", cis_rows)
        print("Significant rows:", len(out_df))

        return {
            "trait": trait,
            "input_file": str(input_file),
            "output_file": str(output_file),
            "total_rows": total_rows,
            "cis_rows": cis_rows,
            "sig_rows": len(out_df),
            "min_p": out_df["P"].min()
        }
    else:
        print("No significant cis variants found.")
        print("Total rows:", total_rows)
        print("Cis rows:", cis_rows)
        print("Significant rows:", sig_rows)

        return {
            "trait": trait,
            "input_file": str(input_file),
            "output_file": None,
            "total_rows": total_rows,
            "cis_rows": cis_rows,
            "sig_rows": 0,
            "min_p": np.nan
        }

sig_selection_records = []

for trait in TARGET_EXPOSURES:
    input_file = SLIM_EXPOSURE_ROOT / trait / f"{trait}_standardised_slim.csv"

    if not input_file.exists():
        print(f"WARNING: Missing slim file for {trait}")
        continue

    record = select_cis_significant_variants_chunked(
        trait=trait,
        input_file=input_file,
        output_dir=SIG_VARIANT_ROOT,
        gene_coordinates=GENE_COORDINATES_GRCH38[trait],
        window=CIS_WINDOW,
        p_threshold=SIG_THRESHOLD,
        chunksize=500000
    )

    sig_selection_records.append(record)

sig_selection_df = pd.DataFrame(sig_selection_records)
sig_selection_path = SUMMARY_DIR / "chunked_cis_significant_variant_selection.csv"
sig_selection_df.to_csv(sig_selection_path, index=False)

print("\nSignificant variant selection summary:")
display(sig_selection_df)
print("Saved to:", sig_selection_path)
```

```python
# ============================================================
# Cell 11: Prepare coordinate-based MR input
# ============================================================

import pandas as pd
import numpy as np
import shutil
from pathlib import Path

MR_READY_SIG_ROOT = BACKEND_ROOT / "mr_ready_significant_variants"
MR_READY_OUTCOME_ROOT = BACKEND_ROOT / "mr_ready_outcome"

if MR_READY_SIG_ROOT.exists():
    shutil.rmtree(MR_READY_SIG_ROOT)
if MR_READY_OUTCOME_ROOT.exists():
    shutil.rmtree(MR_READY_OUTCOME_ROOT)

MR_READY_SIG_ROOT.mkdir(parents=True, exist_ok=True)
MR_READY_OUTCOME_ROOT.mkdir(parents=True, exist_ok=True)

def make_coord_snp_id(df):
    """
    Use CHR:POS as the variant matching key.
    This is safer than CHR:POS:EA:OA when exposure/outcome alleles may be reversed.
    """
    chr_clean = pd.to_numeric(df["CHR"], errors="coerce").astype("Int64").astype(str)
    pos_clean = pd.to_numeric(df["POS"], errors="coerce").astype("Int64").astype(str)
    return chr_clean + ":" + pos_clean

def prepare_mr_ready_exposure_file(input_file: Path, output_file: Path):
    df = pd.read_csv(input_file)

    df["original_SNP"] = df["SNP"].astype(str)
    df["SNP"] = make_coord_snp_id(df)

    # Keep MR-required columns
    keep_cols = ["SNP", "CHR", "POS", "EA", "OA", "BETA", "SE", "EAF", "P", "N", "original_SNP"]
    available_cols = [c for c in keep_cols if c in df.columns]
    df = df[available_cols].copy()

    # Basic cleaning
    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.dropna(subset=["SNP", "EA", "OA", "BETA", "SE", "P"])
    df = df.drop_duplicates(subset=["SNP"], keep="first")

    output_file.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(output_file, index=False)

    return {
        "input_file": str(input_file),
        "output_file": str(output_file),
        "rows": len(df),
        "min_p": df["P"].min() if len(df) > 0 else np.nan,
        "example_snp": df["SNP"].iloc[0] if len(df) > 0 else None,
    }

mr_ready_exposure_records = []

sig_files = sorted(SIG_VARIANT_ROOT.glob("*.csv"))

for sig_file in sig_files:
    trait = sig_file.name.replace("_full_sig_variants.csv", "")
    output_file = MR_READY_SIG_ROOT / f"{trait}_mr_ready_sig_variants.csv"

    record = prepare_mr_ready_exposure_file(sig_file, output_file)
    record["trait"] = trait
    mr_ready_exposure_records.append(record)

mr_ready_exposure_df = pd.DataFrame(mr_ready_exposure_records)

print("MR-ready exposure significant variant files:")
display(mr_ready_exposure_df)

mr_ready_exposure_summary_path = SUMMARY_DIR / "mr_ready_exposure_sig_variant_summary.csv"
mr_ready_exposure_df.to_csv(mr_ready_exposure_summary_path, index=False)

print("Saved to:", mr_ready_exposure_summary_path)
```

```python
# ============================================================
# Cell 12: Prepare coordinate-based outcome file
# ============================================================

def prepare_mr_ready_outcome_file(input_file: Path, output_file: Path, chunksize=500000):
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_chunk = True
    total_rows = 0

    for chunk in pd.read_csv(input_file, chunksize=chunksize, low_memory=False):
        chunk["original_SNP"] = chunk["SNP"].astype(str)
        chunk["SNP"] = make_coord_snp_id(chunk)

        keep_cols = ["SNP", "CHR", "POS", "EA", "OA", "BETA", "SE", "EAF", "P", "N", "original_SNP"]
        available_cols = [c for c in keep_cols if c in chunk.columns]
        out = chunk[available_cols].copy()

        out = out.replace([np.inf, -np.inf], np.nan)
        out = out.dropna(subset=["SNP", "EA", "OA", "BETA", "SE", "P"])

        out.to_csv(
            output_file,
            index=False,
            mode="w" if first_chunk else "a",
            header=first_chunk
        )

        total_rows += len(out)
        first_chunk = False

    return total_rows

outcome_slim_file = SLIM_OUTCOME_ROOT / "endometrial_cancer_standardised_slim.csv"
mr_ready_outcome_file = MR_READY_OUTCOME_ROOT / "endometrial_cancer_mr_ready.csv"

rows = prepare_mr_ready_outcome_file(
    input_file=outcome_slim_file,
    output_file=mr_ready_outcome_file,
    chunksize=500000
)

print("MR-ready outcome file created:")
print("Path:", mr_ready_outcome_file)
print("Rows:", rows)
print("Size MB:", round(mr_ready_outcome_file.stat().st_size / 1024 / 1024, 2))
```

```python
# ============================================================
# Cell 14: Import R packages for MR analysis
# ============================================================

import rpy2.robjects as robjects
from rpy2.robjects.packages import importr
from rpy2.robjects import pandas2ri
from rpy2.rinterface_lib.embedded import RRuntimeError

pandas2ri.activate()

TwoSampleMR = importr("TwoSampleMR")
grdevices = importr("grDevices")

rprint = robjects.r("function(x, ...) print(x)")

def r_to_pandas(r_obj):
    with (robjects.default_converter + pandas2ri.converter).context():
        return robjects.conversion.get_conversion().rpy2py(r_obj)

print("TwoSampleMR imported successfully through rpy2.")
```

```python
# ============================================================
# Cell 15: Run coordinate-based MR without LD clumping
# ============================================================

MR_COORD_OUTPUT_ROOT = MR_OUTPUT_ROOT / "coordinate_based_no_ld_clump"

COORD_RESULTS_DIR = MR_COORD_OUTPUT_ROOT / "results"
COORD_HARMONISED_DIR = MR_COORD_OUTPUT_ROOT / "harmonised_data"
COORD_PLEIOTROPY_DIR = MR_COORD_OUTPUT_ROOT / "pleiotropy"
COORD_HETEROGENEITY_DIR = MR_COORD_OUTPUT_ROOT / "heterogeneity"
COORD_SCATTER_DIR = MR_COORD_OUTPUT_ROOT / "scatter_plots"
COORD_SINGLE_SNP_DIR = MR_COORD_OUTPUT_ROOT / "singleSNP_plots"

for folder in [
    COORD_RESULTS_DIR,
    COORD_HARMONISED_DIR,
    COORD_PLEIOTROPY_DIR,
    COORD_HETEROGENEITY_DIR,
    COORD_SCATTER_DIR,
    COORD_SINGLE_SNP_DIR,
]:
    folder.mkdir(parents=True, exist_ok=True)

mr_run_records = []
mr_result_tables = []

for _, row in mr_ready_exposure_df.iterrows():
    trait = row["trait"]
    exp_file = Path(row["output_file"])

    print("\n" + "=" * 100)
    print("Running MR for:", trait)
    print("Exposure file:", exp_file)
    print("Outcome file:", mr_ready_outcome_file)

    try:
        # Read exposure significant variants
        exp_dat = TwoSampleMR.read_exposure_data(
            filename=str(exp_file),
            sep=",",
            snp_col="SNP",
            beta_col="BETA",
            se_col="SE",
            effect_allele_col="EA",
            other_allele_col="OA",
            eaf_col="EAF",
            pval_col="P",
            samplesize_col="N",
            chr_col="CHR",
            pos_col="POS",
            clump=False
        )

        exp_df = pd.read_csv(exp_file)
        snp_list = robjects.StrVector(exp_df["SNP"].dropna().astype(str).tolist())

        # Read matching outcome rows
        outcome_dat = TwoSampleMR.read_outcome_data(
            snps=snp_list,
            filename=str(mr_ready_outcome_file),
            sep=",",
            snp_col="SNP",
            beta_col="BETA",
            se_col="SE",
            effect_allele_col="EA",
            other_allele_col="OA",
            eaf_col="EAF",
            pval_col="P"
        )

        # Harmonise
        harmonised = TwoSampleMR.harmonise_data(exp_dat, outcome_dat)
        harmonised_py = r_to_pandas(harmonised)

        print("Harmonised SNP count:", len(harmonised_py))

        if len(harmonised_py) == 0:
            raise ValueError("No SNPs retained after harmonisation.")

        # Run MR
        res = TwoSampleMR.mr(harmonised)
        res_py = r_to_pandas(res)

        # Sensitivity tests
        try:
            pleiotropy = TwoSampleMR.mr_pleiotropy_test(harmonised)
            pleiotropy_py = r_to_pandas(pleiotropy)
        except Exception as e:
            pleiotropy_py = pd.DataFrame({"warning": [str(e)]})

        try:
            heterogeneity = TwoSampleMR.mr_heterogeneity(harmonised)
            heterogeneity_py = r_to_pandas(heterogeneity)
        except Exception as e:
            heterogeneity_py = pd.DataFrame({"warning": [str(e)]})

        # Single SNP
        try:
            single_snp = TwoSampleMR.mr_singlesnp(harmonised)
        except Exception as e:
            single_snp = None
            print("Single-SNP warning:", e)

        # Save tables
        res_py.to_csv(COORD_RESULTS_DIR / f"{trait}.csv", index=False)
        harmonised_py.to_csv(COORD_HARMONISED_DIR / f"{trait}.csv", index=False)
        pleiotropy_py.to_csv(COORD_PLEIOTROPY_DIR / f"{trait}.csv", index=False)
        heterogeneity_py.to_csv(COORD_HETEROGENEITY_DIR / f"{trait}.csv", index=False)

        # Save scatter plot
        try:
            scatter_plot = TwoSampleMR.mr_scatter_plot(res, harmonised)
            grdevices.png(
                file=str(COORD_SCATTER_DIR / f"{trait}_scatter.png"),
                width=1000,
                height=850
            )
            rprint(scatter_plot)
            grdevices.dev_off()
        except Exception as e:
            print("Scatter plot warning:", e)

        # Save single SNP plot
        if single_snp is not None:
            try:
                forest_plot = TwoSampleMR.mr_forest_plot(single_snp)
                grdevices.png(
                    file=str(COORD_SINGLE_SNP_DIR / f"{trait}_singleSNP.png"),
                    width=1000,
                    height=850
                )
                rprint(forest_plot)
                grdevices.dev_off()
            except Exception as e:
                print("Single-SNP plot warning:", e)

        res_py["trait"] = trait
        mr_result_tables.append(res_py)

        mr_run_records.append({
            "trait": trait,
            "status": "success",
            "harmonised_snps": len(harmonised_py),
            "result_rows": len(res_py),
            "error": None
        })

        print("MR result:")
        display(res_py)

    except Exception as e:
        print("ERROR:", e)

        mr_run_records.append({
            "trait": trait,
            "status": "failed",
            "harmonised_snps": 0,
            "result_rows": 0,
            "error": str(e)
        })

mr_run_summary = pd.DataFrame(mr_run_records)
mr_run_summary_path = SUMMARY_DIR / "coordinate_based_mr_run_summary.csv"
mr_run_summary.to_csv(mr_run_summary_path, index=False)

print("\nMR run summary:")
display(mr_run_summary)
print("Saved to:", mr_run_summary_path)
```

```python
# ============================================================
# Cell 16: Combine coordinate-based MR results
# ============================================================

if len(mr_result_tables) > 0:
    combined_coord_mr_results = pd.concat(mr_result_tables, ignore_index=True)

    # Add OR interpretation
    if "b" in combined_coord_mr_results.columns:
        combined_coord_mr_results["MR_OR"] = np.exp(combined_coord_mr_results["b"])

    if "b" in combined_coord_mr_results.columns and "se" in combined_coord_mr_results.columns:
        combined_coord_mr_results["MR_OR_lower_95CI"] = np.exp(
            combined_coord_mr_results["b"] - 1.96 * combined_coord_mr_results["se"]
        )
        combined_coord_mr_results["MR_OR_upper_95CI"] = np.exp(
            combined_coord_mr_results["b"] + 1.96 * combined_coord_mr_results["se"]
        )

    if "pval" in combined_coord_mr_results.columns:
        combined_coord_mr_results["nominal_significant"] = combined_coord_mr_results["pval"] < 0.05

    combined_coord_results_path = SUMMARY_DIR / "coordinate_based_combined_mr_results.csv"
    combined_coord_mr_results.to_csv(combined_coord_results_path, index=False)

    print("Combined coordinate-based MR results:")
    display(combined_coord_mr_results)
    print("Saved to:", combined_coord_results_path)

else:
    print("No MR result tables were generated.")
```

```python
# ============================================================
# Cell 18: Prepare multiple outcome files for comparison
# ============================================================

import pandas as pd
import numpy as np
import shutil
from pathlib import Path

OUTCOME_COMPARE_ROOT = MR_OUTPUT_ROOT / "outcome_comparison"
OUTCOME_COMPARE_PREPARED_ROOT = OUTCOME_COMPARE_ROOT / "prepared_outcomes"
OUTCOME_COMPARE_MR_READY_ROOT = OUTCOME_COMPARE_ROOT / "mr_ready_outcomes"

for folder in [
    OUTCOME_COMPARE_ROOT,
    OUTCOME_COMPARE_PREPARED_ROOT,
    OUTCOME_COMPARE_MR_READY_ROOT,
]:
    folder.mkdir(parents=True, exist_ok=True)

def prepare_single_outcome_for_comparison(set_info):
    set_name = set_info["set_name"]
    outcome_input_file = Path(set_info["outcome_file"])

    print("\n" + "=" * 100)
    print("Preparing outcome set:", set_name)
    print("Outcome file:", outcome_input_file)

    prepared_output_file = OUTCOME_COMPARE_PREPARED_ROOT / set_name / "endometrial_cancer_standardised.csv"
    mr_ready_output_file = OUTCOME_COMPARE_MR_READY_ROOT / set_name / "endometrial_cancer_mr_ready.csv"

    # Standardise outcome file.
    # This uses your existing standardise_gwas_file function.
    record = standardise_gwas_file(
        input_path=outcome_input_file,
        output_path=prepared_output_file,
        trait=TARGET_OUTCOME,
        role="outcome",
        force_binary_outcome_or_to_beta=True
    )

    # Convert to coordinate-based MR-ready outcome file.
    rows = prepare_mr_ready_outcome_file(
        input_file=prepared_output_file,
        output_file=mr_ready_output_file,
        chunksize=500000
    )

    return {
        "set_name": set_name,
        "outcome_file": str(outcome_input_file),
        "prepared_outcome_file": str(prepared_output_file),
        "mr_ready_outcome_file": str(mr_ready_output_file),
        "rows_saved": record.get("rows_saved", None),
        "effect_size_type": record.get("effect_size_type", None),
        "snp_source_type": record.get("snp_source_type", None),
        "build": record.get("build", None),
        "mr_ready_rows": rows,
        "status": "prepared",
        "error": None,
    }

multi_outcome_prepare_records = []

for set_info in MULTI_OUTCOME_ANALYSIS_SETS:
    try:
        record = prepare_single_outcome_for_comparison(set_info)
        multi_outcome_prepare_records.append(record)
    except Exception as e:
        print("ERROR:", e)
        multi_outcome_prepare_records.append({
            "set_name": set_info["set_name"],
            "outcome_file": set_info["outcome_file"],
            "prepared_outcome_file": None,
            "mr_ready_outcome_file": None,
            "rows_saved": 0,
            "effect_size_type": None,
            "snp_source_type": None,
            "build": set_info.get("outcome_build", None),
            "mr_ready_rows": 0,
            "status": "failed",
            "error": str(e),
        })

multi_outcome_prepare_df = pd.DataFrame(multi_outcome_prepare_records)

multi_outcome_prepare_path = SUMMARY_DIR / "multi_outcome_prepare_summary.csv"
multi_outcome_prepare_df.to_csv(multi_outcome_prepare_path, index=False)

print("\nMulti-outcome preparation summary:")
display(multi_outcome_prepare_df)
print("Saved to:", multi_outcome_prepare_path)
```

```python
# ============================================================
# Cell 19: Check SNP overlap for each outcome candidate
# ============================================================

def load_snp_set_from_csv(path: Path, snp_col="SNP", chunksize=500000):
    snps = set()
    for chunk in pd.read_csv(path, usecols=[snp_col], chunksize=chunksize):
        snps.update(chunk[snp_col].dropna().astype(str).tolist())
    return snps

multi_outcome_overlap_records = []

# Reuse MR-ready exposure significant variant files from previous cells
mr_ready_sig_files = sorted(MR_READY_SIG_ROOT.glob("*_mr_ready_sig_variants.csv"))

for _, outcome_row in multi_outcome_prepare_df.iterrows():
    if outcome_row["status"] != "prepared":
        continue

    set_name = outcome_row["set_name"]
    outcome_file = Path(outcome_row["mr_ready_outcome_file"])

    print("\n" + "=" * 100)
    print("Checking overlap for:", set_name)

    outcome_snps = load_snp_set_from_csv(outcome_file)

    for exp_file in mr_ready_sig_files:
        trait = exp_file.name.replace("_mr_ready_sig_variants.csv", "")
        exp_df = pd.read_csv(exp_file)
        exp_snps = set(exp_df["SNP"].dropna().astype(str).tolist())

        overlap = sorted(list(exp_snps.intersection(outcome_snps)))

        multi_outcome_overlap_records.append({
            "set_name": set_name,
            "trait": trait,
            "exposure_snps": len(exp_snps),
            "outcome_snps": len(outcome_snps),
            "overlap_snps": len(overlap),
            "example_overlap": overlap[0] if len(overlap) > 0 else None,
            "outcome_file": str(outcome_file),
            "exposure_file": str(exp_file),
        })

multi_outcome_overlap_df = pd.DataFrame(multi_outcome_overlap_records)

multi_outcome_overlap_path = SUMMARY_DIR / "multi_outcome_snp_overlap.csv"
multi_outcome_overlap_df.to_csv(multi_outcome_overlap_path, index=False)

print("Multi-outcome SNP overlap:")
display(multi_outcome_overlap_df)
print("Saved to:", multi_outcome_overlap_path)
```

```python
# ============================================================
# Cell 20: Run MR for each outcome candidate
# ============================================================

MULTI_OUTCOME_MR_ROOT = OUTCOME_COMPARE_ROOT / "mr_results"

MULTI_OUTCOME_RESULTS_DIR = MULTI_OUTCOME_MR_ROOT / "results"
MULTI_OUTCOME_HARMONISED_DIR = MULTI_OUTCOME_MR_ROOT / "harmonised_data"
MULTI_OUTCOME_PLEIOTROPY_DIR = MULTI_OUTCOME_MR_ROOT / "pleiotropy"
MULTI_OUTCOME_HETEROGENEITY_DIR = MULTI_OUTCOME_MR_ROOT / "heterogeneity"

for folder in [
    MULTI_OUTCOME_RESULTS_DIR,
    MULTI_OUTCOME_HARMONISED_DIR,
    MULTI_OUTCOME_PLEIOTROPY_DIR,
    MULTI_OUTCOME_HETEROGENEITY_DIR,
]:
    folder.mkdir(parents=True, exist_ok=True)

multi_outcome_mr_run_records = []
multi_outcome_result_tables = []

for _, outcome_row in multi_outcome_prepare_df.iterrows():
    if outcome_row["status"] != "prepared":
        continue

    set_name = outcome_row["set_name"]
    outcome_file = Path(outcome_row["mr_ready_outcome_file"])

    print("\n" + "#" * 120)
    print("Running MR for outcome set:", set_name)
    print("Outcome file:", outcome_file)

    for exp_file in mr_ready_sig_files:
        trait = exp_file.name.replace("_mr_ready_sig_variants.csv", "")

        print("\n" + "=" * 100)
        print("Exposure:", trait)
        print("Exposure file:", exp_file)

        try:
            exp_dat = TwoSampleMR.read_exposure_data(
                filename=str(exp_file),
                sep=",",
                snp_col="SNP",
                beta_col="BETA",
                se_col="SE",
                effect_allele_col="EA",
                other_allele_col="OA",
                eaf_col="EAF",
                pval_col="P",
                samplesize_col="N",
                chr_col="CHR",
                pos_col="POS",
                clump=False
            )

            exp_df = pd.read_csv(exp_file)
            snp_list = robjects.StrVector(exp_df["SNP"].dropna().astype(str).tolist())

            outcome_dat = TwoSampleMR.read_outcome_data(
                snps=snp_list,
                filename=str(outcome_file),
                sep=",",
                snp_col="SNP",
                beta_col="BETA",
                se_col="SE",
                effect_allele_col="EA",
                other_allele_col="OA",
                eaf_col="EAF",
                pval_col="P"
            )

            harmonised = TwoSampleMR.harmonise_data(exp_dat, outcome_dat)
            harmonised_py = r_to_pandas(harmonised)

            harmonised_snps = len(harmonised_py)
            print("Harmonised SNP count:", harmonised_snps)

            if harmonised_snps == 0:
                raise ValueError("No SNPs retained after harmonisation.")

            res = TwoSampleMR.mr(harmonised)
            res_py = r_to_pandas(res)

            try:
                pleiotropy = TwoSampleMR.mr_pleiotropy_test(harmonised)
                pleiotropy_py = r_to_pandas(pleiotropy)
            except Exception as e:
                pleiotropy_py = pd.DataFrame({"warning": [str(e)]})

            try:
                heterogeneity = TwoSampleMR.mr_heterogeneity(harmonised)
                heterogeneity_py = r_to_pandas(heterogeneity)
            except Exception as e:
                heterogeneity_py = pd.DataFrame({"warning": [str(e)]})

            # Add metadata
            res_py["set_name"] = set_name
            res_py["trait"] = trait
            res_py["outcome_file"] = str(outcome_file)
            res_py["harmonised_snps"] = harmonised_snps

            # Save outputs
            safe_name = f"{set_name}_{trait}"
            res_py.to_csv(MULTI_OUTCOME_RESULTS_DIR / f"{safe_name}.csv", index=False)
            harmonised_py.to_csv(MULTI_OUTCOME_HARMONISED_DIR / f"{safe_name}.csv", index=False)
            pleiotropy_py.to_csv(MULTI_OUTCOME_PLEIOTROPY_DIR / f"{safe_name}.csv", index=False)
            heterogeneity_py.to_csv(MULTI_OUTCOME_HETEROGENEITY_DIR / f"{safe_name}.csv", index=False)

            multi_outcome_result_tables.append(res_py)

            multi_outcome_mr_run_records.append({
                "set_name": set_name,
                "trait": trait,
                "status": "success",
                "harmonised_snps": harmonised_snps,
                "result_rows": len(res_py),
                "error": None,
            })

            display(res_py)

        except Exception as e:
            print("ERROR:", e)

            multi_outcome_mr_run_records.append({
                "set_name": set_name,
                "trait": trait,
                "status": "failed",
                "harmonised_snps": 0,
                "result_rows": 0,
                "error": str(e),
            })

multi_outcome_mr_run_df = pd.DataFrame(multi_outcome_mr_run_records)

multi_outcome_mr_run_path = SUMMARY_DIR / "multi_outcome_mr_run_summary.csv"
multi_outcome_mr_run_df.to_csv(multi_outcome_mr_run_path, index=False)

print("\nMulti-outcome MR run summary:")
display(multi_outcome_mr_run_df)
print("Saved to:", multi_outcome_mr_run_path)
```

```python
# ============================================================
# Cell 21: Combine and rank multi-outcome MR results
# ============================================================

if len(multi_outcome_result_tables) > 0:
    multi_outcome_combined_results = pd.concat(multi_outcome_result_tables, ignore_index=True)

    if "b" in multi_outcome_combined_results.columns:
        multi_outcome_combined_results["MR_OR"] = np.exp(multi_outcome_combined_results["b"])

    if "b" in multi_outcome_combined_results.columns and "se" in multi_outcome_combined_results.columns:
        multi_outcome_combined_results["MR_OR_lower_95CI"] = np.exp(
            multi_outcome_combined_results["b"] - 1.96 * multi_outcome_combined_results["se"]
        )
        multi_outcome_combined_results["MR_OR_upper_95CI"] = np.exp(
            multi_outcome_combined_results["b"] + 1.96 * multi_outcome_combined_results["se"]
        )

    if "pval" in multi_outcome_combined_results.columns:
        multi_outcome_combined_results["nominal_significant"] = multi_outcome_combined_results["pval"] < 0.05

    # Merge overlap info
    overlap_for_merge = multi_outcome_overlap_df[[
        "set_name", "trait", "overlap_snps"
    ]].drop_duplicates()

    multi_outcome_combined_results = multi_outcome_combined_results.merge(
        overlap_for_merge,
        on=["set_name", "trait"],
        how="left"
    )

    # Sort by trait, p-value, and overlap
    if "pval" in multi_outcome_combined_results.columns:
        multi_outcome_combined_results = multi_outcome_combined_results.sort_values(
            by=["trait", "pval", "overlap_snps"],
            ascending=[True, True, False]
        )

    multi_outcome_combined_path = SUMMARY_DIR / "multi_outcome_combined_mr_results.csv"
    multi_outcome_combined_results.to_csv(multi_outcome_combined_path, index=False)

    print("Multi-outcome combined MR results:")
    display(multi_outcome_combined_results)
    print("Saved to:", multi_outcome_combined_path)

else:
    multi_outcome_combined_results = pd.DataFrame()
    print("No multi-outcome MR result tables were generated.")
```

```python
# ============================================================
# Cell 23.2: rsID mapping helper using Ensembl REST
# ============================================================

import requests
import time

ENSEMBL_REST_SERVER = "https://rest.ensembl.org"

def query_ensembl_variants_by_position(chrom, pos, max_retries=3, sleep_seconds=0.15):
    """
    Query Ensembl REST API for variants overlapping one GRCh38 position.
    Returns a list of variant records.
    """
    chrom = str(chrom).replace("chr", "").strip()
    pos = int(pos)

    url = f"{ENSEMBL_REST_SERVER}/overlap/region/human/{chrom}:{pos}-{pos}"
    params = {"feature": "variation"}

    headers = {
        "Content-Type": "application/json",
        "Accept": "application/json"
    }

    for attempt in range(max_retries):
        try:
            r = requests.get(url, headers=headers, params=params, timeout=20)

            if r.status_code == 200:
                data = r.json()
                if isinstance(data, list):
                    return data
                return []

            if r.status_code in [429, 500, 502, 503, 504]:
                time.sleep(1 + attempt)
                continue

            return []

        except Exception:
            time.sleep(1 + attempt)

    return []


def pick_best_rsid_from_ensembl_records(records):
    """
    Pick the first rsID-like variant ID from Ensembl overlap records.
    """
    if records is None or len(records) == 0:
        return None

    rsids = []

    for rec in records:
        vid = str(rec.get("id", ""))
        if vid.startswith("rs"):
            rsids.append(vid)

    if len(rsids) > 0:
        return rsids[0]

    return None


def map_single_variant_to_rsid(row):
    """
    Map one row with CHR and POS to rsID.
    If the original SNP is already rsID, keep it.
    """
    original_snp = str(row.get("SNP", ""))

    if original_snp.startswith("rs"):
        return {
            "mapped_rsid": original_snp,
            "mapping_status": "already_rsid",
            "ensembl_record_count": np.nan
        }

    chrom = row.get("CHR", None)
    pos = row.get("POS", None)

    if pd.isna(chrom) or pd.isna(pos):
        return {
            "mapped_rsid": None,
            "mapping_status": "missing_chr_pos",
            "ensembl_record_count": 0
        }

    records = query_ensembl_variants_by_position(chrom, pos)
    mapped_rsid = pick_best_rsid_from_ensembl_records(records)

    if mapped_rsid is not None:
        return {
            "mapped_rsid": mapped_rsid,
            "mapping_status": "mapped_by_chr_pos",
            "ensembl_record_count": len(records)
        }

    return {
        "mapped_rsid": None,
        "mapping_status": "not_mapped",
        "ensembl_record_count": len(records)
    }

print("rsID mapping helper is ready.")
```

```python
# ============================================================
# Cell 23.3: Map significant variants to rsID
# ============================================================

mapped_variant_records = []
mapped_file_records = []

for sig_file in sig_files:
    trait = sig_file.name.replace("_full_sig_variants.csv", "")
    df = pd.read_csv(sig_file)

    required_cols = ["SNP", "CHR", "POS", "EA", "OA", "BETA", "SE", "EAF", "P", "N"]
    for col in required_cols:
        if col not in df.columns:
            df[col] = pd.NA

    print("\n" + "=" * 100)
    print("Mapping rsID for:", trait)
    print("Input:", sig_file)
    print("Rows:", len(df))

    mapped_rows = []

    for idx, row in df.iterrows():
        mapping = map_single_variant_to_rsid(row)

        new_row = row.to_dict()
        new_row["trait"] = trait
        new_row["original_SNP"] = str(row.get("SNP", ""))
        new_row["mapped_rsid"] = mapping["mapped_rsid"]
        new_row["mapping_status"] = mapping["mapping_status"]
        new_row["ensembl_record_count"] = mapping["ensembl_record_count"]

        mapped_rows.append(new_row)

        if (idx + 1) % 20 == 0:
            print(f"Mapped {idx + 1}/{len(df)} variants...")

    mapped_df = pd.DataFrame(mapped_rows)

    mapped_output_file = LD_MAPPING_ROOT / f"{trait}_significant_variants_rsid_mapping.csv"
    mapped_df.to_csv(mapped_output_file, index=False)

    mapped_count = int(mapped_df["mapped_rsid"].notna().sum())
    already_rsid_count = int((mapped_df["mapping_status"] == "already_rsid").sum())
    mapped_by_pos_count = int((mapped_df["mapping_status"] == "mapped_by_chr_pos").sum())
    not_mapped_count = int((mapped_df["mapping_status"] == "not_mapped").sum())

    mapped_file_records.append({
        "trait": trait,
        "input_file": str(sig_file),
        "mapped_output_file": str(mapped_output_file),
        "total_variants": len(mapped_df),
        "mapped_rsid_count": mapped_count,
        "already_rsid_count": already_rsid_count,
        "mapped_by_chr_pos_count": mapped_by_pos_count,
        "not_mapped_count": not_mapped_count,
        "mapping_rate": mapped_count / len(mapped_df) if len(mapped_df) > 0 else 0
    })

    print("Saved:", mapped_output_file)
    print("Mapped rsID:", mapped_count, "/", len(mapped_df))

rsid_mapping_summary = pd.DataFrame(mapped_file_records)

rsid_mapping_summary_path = SUMMARY_DIR / "rsid_mapping_summary.csv"
rsid_mapping_summary.to_csv(rsid_mapping_summary_path, index=False)

print("\nrsID mapping summary:")
display(rsid_mapping_summary)
print("Saved to:", rsid_mapping_summary_path)
```

```python
# ============================================================
# Cell 23.7.0
# Recovery cell before Cell 23.7:
# Define helper function for loading SNP sets from CSV
# ============================================================

import pandas as pd
from pathlib import Path

def load_snp_set_from_csv(file_path, snp_col="SNP", chunksize=500000):
    """
    Load SNP identifiers from a large CSV file safely using chunks.
    This is used to check overlap between exposure SNPs and outcome SNPs.
    """
    file_path = Path(file_path)

    if not file_path.exists():
        raise FileNotFoundError(f"File not found: {file_path}")

    snp_set = set()

    for chunk in pd.read_csv(file_path, usecols=[snp_col], chunksize=chunksize):
        snps = chunk[snp_col].dropna().astype(str).tolist()
        snp_set.update(snps)

    return snp_set

print("load_snp_set_from_csv helper is ready.")
```

```python
# ============================================================
# Cell 23.7: Check clumped exposure-outcome overlap for multiple outcomes
# ============================================================

multi_outcome_prepare_path = SUMMARY_DIR / "multi_outcome_prepare_summary.csv"

if "multi_outcome_prepare_df" not in globals():
    if multi_outcome_prepare_path.exists():
        multi_outcome_prepare_df = pd.read_csv(multi_outcome_prepare_path)
        print("Loaded multi_outcome_prepare_df from:", multi_outcome_prepare_path)
    else:
        raise FileNotFoundError(
            "multi_outcome_prepare_df is not in memory and multi_outcome_prepare_summary.csv was not found. "
            "Please run Cell 18 first."
        )

clumped_sig_files = sorted(MR_READY_CLUMPED_SIG_ROOT.glob("*_mr_ready_clumped_sig_variants.csv"))

if len(clumped_sig_files) == 0:
    raise FileNotFoundError(
        f"No MR-ready clumped significant variant files found in {MR_READY_CLUMPED_SIG_ROOT}. "
        "Please complete Cell 23.5 and Cell 23.6 first."
    )

clumped_overlap_records = []

for _, outcome_row in multi_outcome_prepare_df.iterrows():
    if str(outcome_row["status"]).lower() != "prepared":
        continue

    set_name = outcome_row["set_name"]
    outcome_file = Path(outcome_row["mr_ready_outcome_file"])

    print("\n" + "=" * 100)
    print("Checking clumped overlap for:", set_name)

    outcome_snps = load_snp_set_from_csv(outcome_file)

    for exp_file in clumped_sig_files:
        trait = exp_file.name.replace("_mr_ready_clumped_sig_variants.csv", "")
        exp_df = pd.read_csv(exp_file)
        exp_snps = set(exp_df["SNP"].dropna().astype(str).tolist())

        overlap = sorted(list(exp_snps.intersection(outcome_snps)))

        clumped_overlap_records.append({
            "set_name": set_name,
            "trait": trait,
            "clumped_exposure_snps": len(exp_snps),
            "outcome_snps": len(outcome_snps),
            "overlap_snps": len(overlap),
            "example_overlap": overlap[0] if len(overlap) > 0 else None,
            "outcome_file": str(outcome_file),
            "clumped_exposure_file": str(exp_file),
        })

clumped_multi_outcome_overlap_df = pd.DataFrame(clumped_overlap_records)

clumped_overlap_path = SUMMARY_DIR / "clumped_multi_outcome_snp_overlap.csv"
clumped_multi_outcome_overlap_df.to_csv(clumped_overlap_path, index=False)

display(clumped_multi_outcome_overlap_df)
print("Saved to:", clumped_overlap_path)
```

```python
# ============================================================
# Cell 23.8: Run clumped MR for each outcome candidate
# ============================================================

clumped_mr_run_records = []
clumped_mr_result_tables = []

for _, outcome_row in multi_outcome_prepare_df.iterrows():
    if str(outcome_row["status"]).lower() != "prepared":
        continue

    set_name = outcome_row["set_name"]
    outcome_file = Path(outcome_row["mr_ready_outcome_file"])

    print("\n" + "#" * 120)
    print("Running clumped MR for outcome set:", set_name)
    print("Outcome file:", outcome_file)

    for exp_file in clumped_sig_files:
        trait = exp_file.name.replace("_mr_ready_clumped_sig_variants.csv", "")

        print("\n" + "=" * 100)
        print("Exposure:", trait)
        print("Clumped exposure file:", exp_file)

        try:
            exp_df = pd.read_csv(exp_file)

            if len(exp_df) == 0:
                raise ValueError("Clumped exposure file is empty.")

            exp_dat = TwoSampleMR.read_exposure_data(
                filename=str(exp_file),
                sep=",",
                snp_col="SNP",
                beta_col="BETA",
                se_col="SE",
                effect_allele_col="EA",
                other_allele_col="OA",
                eaf_col="EAF",
                pval_col="P",
                samplesize_col="N",
                chr_col="CHR",
                pos_col="POS",
                clump=False
            )

            snp_list = robjects.StrVector(exp_df["SNP"].dropna().astype(str).tolist())

            outcome_dat = TwoSampleMR.read_outcome_data(
                snps=snp_list,
                filename=str(outcome_file),
                sep=",",
                snp_col="SNP",
                beta_col="BETA",
                se_col="SE",
                effect_allele_col="EA",
                other_allele_col="OA",
                eaf_col="EAF",
                pval_col="P"
            )

            harmonised = TwoSampleMR.harmonise_data(exp_dat, outcome_dat)
            harmonised_py = r_to_pandas(harmonised)

            harmonised_snps = len(harmonised_py)
            print("Harmonised SNP count:", harmonised_snps)

            if harmonised_snps == 0:
                raise ValueError("No SNPs retained after harmonisation.")

            res = TwoSampleMR.mr(harmonised)
            res_py = r_to_pandas(res)

            try:
                pleiotropy = TwoSampleMR.mr_pleiotropy_test(harmonised)
                pleiotropy_py = r_to_pandas(pleiotropy)
            except Exception as e:
                pleiotropy_py = pd.DataFrame({"warning": [str(e)]})

            try:
                heterogeneity = TwoSampleMR.mr_heterogeneity(harmonised)
                heterogeneity_py = r_to_pandas(heterogeneity)
            except Exception as e:
                heterogeneity_py = pd.DataFrame({"warning": [str(e)]})

            res_py["set_name"] = set_name
            res_py["trait"] = trait
            res_py["outcome_file"] = str(outcome_file)
            res_py["harmonised_snps"] = harmonised_snps
            res_py["instrument_selection"] = "LD clumped"

            safe_name = f"{set_name}_{trait}_ld_clumped"

            res_py.to_csv(CLUMPED_MR_RESULTS_DIR / f"{safe_name}.csv", index=False)
            harmonised_py.to_csv(CLUMPED_MR_HARMONISED_DIR / f"{safe_name}.csv", index=False)
            pleiotropy_py.to_csv(CLUMPED_MR_PLEIOTROPY_DIR / f"{safe_name}.csv", index=False)
            heterogeneity_py.to_csv(CLUMPED_MR_HETEROGENEITY_DIR / f"{safe_name}.csv", index=False)

            clumped_mr_result_tables.append(res_py)

            clumped_mr_run_records.append({
                "set_name": set_name,
                "trait": trait,
                "status": "success",
                "harmonised_snps": harmonised_snps,
                "result_rows": len(res_py),
                "error": None,
            })

            display(res_py)

        except Exception as e:
            print("ERROR:", e)

            clumped_mr_run_records.append({
                "set_name": set_name,
                "trait": trait,
                "status": "failed",
                "harmonised_snps": 0,
                "result_rows": 0,
                "error": str(e),
            })

clumped_mr_run_df = pd.DataFrame(clumped_mr_run_records)

clumped_mr_run_path = SUMMARY_DIR / "clumped_multi_outcome_mr_run_summary.csv"
clumped_mr_run_df.to_csv(clumped_mr_run_path, index=False)

display(clumped_mr_run_df)
print("Saved to:", clumped_mr_run_path)
```

```python
# ============================================================
# Cell 23.9: Combine clumped MR results
# ============================================================

if len(clumped_mr_result_tables) > 0:
    clumped_combined_mr_results = pd.concat(clumped_mr_result_tables, ignore_index=True)

    if "b" in clumped_combined_mr_results.columns:
        clumped_combined_mr_results["MR_OR"] = np.exp(clumped_combined_mr_results["b"])

    if "b" in clumped_combined_mr_results.columns and "se" in clumped_combined_mr_results.columns:
        clumped_combined_mr_results["MR_OR_lower_95CI"] = np.exp(
            clumped_combined_mr_results["b"] - 1.96 * clumped_combined_mr_results["se"]
        )
        clumped_combined_mr_results["MR_OR_upper_95CI"] = np.exp(
            clumped_combined_mr_results["b"] + 1.96 * clumped_combined_mr_results["se"]
        )

    if "pval" in clumped_combined_mr_results.columns:
        clumped_combined_mr_results["nominal_significant"] = clumped_combined_mr_results["pval"] < 0.05

    if "clumped_multi_outcome_overlap_df" in globals() and len(clumped_multi_outcome_overlap_df) > 0:
        overlap_for_merge = clumped_multi_outcome_overlap_df[[
            "set_name", "trait", "overlap_snps"
        ]].drop_duplicates()

        clumped_combined_mr_results = clumped_combined_mr_results.merge(
            overlap_for_merge,
            on=["set_name", "trait"],
            how="left"
        )

    if "pval" in clumped_combined_mr_results.columns:
        clumped_combined_mr_results = clumped_combined_mr_results.sort_values(
            by=["trait", "pval"],
            ascending=[True, True]
        )

    clumped_combined_path = SUMMARY_DIR / "clumped_multi_outcome_combined_mr_results.csv"
    clumped_combined_mr_results.to_csv(clumped_combined_path, index=False)

    display(clumped_combined_mr_results)
    print("Saved to:", clumped_combined_path)

else:
    clumped_combined_mr_results = pd.DataFrame()
    print("No clumped MR result tables were generated.")
```
